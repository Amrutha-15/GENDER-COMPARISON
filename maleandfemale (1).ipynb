```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data: female and male weights
female_weights = np.random.normal(65, 10, 500)  # 500 data points, mean=65, std=10
male_weights = np.random.normal(80, 15, 500)  # 500 data points, mean=80, std=15

# Create the figure and subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 6))

# Plot the female weights histogram (top subplot)
ax1.hist(female_weights, bins=20, color='pink', edgecolor='black')
ax1.set_title("Female Weights")
ax1.set_xlabel("Weight (kg)")
ax1.set_ylabel("Frequency")

# Plot the male weights histogram (bottom subplot)
ax2.hist(male_weights, bins=20, color='blue', edgecolor='black')
ax2.set_title("Male Weights")
ax2.set_xlabel("Weight (kg)")
ax2.set_ylabel("Frequency")

# Set identical x-axis limits for both subplots
min_weight = min(female_weights.min(), male_weights.min())
max_weight = max(female_weights.max(), male_weights.max())
ax1.set_xlim(min_weight, max_weight)
ax2.set_xlim(min_weight, max_weight)

# Adjust layout for better spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import numpy as np
import pandas as pd

# Sample data: female weights
female_weights = np.random.normal(65, 10, 500)  # 500 data points, mean=65, std=10

# Generate random height data for female participants (in meters)
female_heights = np.random.normal(1.65, 0.1, 500)  # Mean height = 1.65m, std = 0.1m

# Calculate BMI using the formula: BMI = weight / (height^2)
female_bmi = female_weights / (female_heights ** 2)

# Combine the data into a matrix
female_data = np.column_stack((female_weights, female_heights, female_bmi))

# Standardize each column to compute the z-scores
zfemale = (female_data - np.mean(female_data, axis=0)) / np.std(female_data, axis=0)

# If you'd like to see the result as a DataFrame for easy inspection
df_zfemale = pd.DataFrame(zfemale, columns=['Weight', 'Height', 'BMI'])

# Display the first few rows of the standardized dataset
print(df_zfemale.head())
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr, spearmanr

# Generate additional data: waist and hip circumference (in cm)
# Assuming waist and hip circumference are somewhat correlated with weight and height
waist_circumference = np.random.normal(80, 10, 500)  # Mean waist circumference = 80 cm, std = 10 cm
hip_circumference = np.random.normal(100, 12, 500)  # Mean hip circumference = 100 cm, std = 12 cm

# Combine the existing standardized data (zfemale) with waist and hip data
# Assuming zfemale is already created with standardized weight, height, and BMI
female_data_full = np.column_stack((female_weights, female_heights, female_bmi, waist_circumference, hip_circumference))

# Standardize waist and hip circumference (using z-scores)
zwaist = (waist_circumference - np.mean(waist_circumference)) / np.std(waist_circumference)
zhip = (hip_circumference - np.mean(hip_circumference)) / np.std(hip_circumference)

# Add the standardized waist and hip data to the zfemale matrix
zfemale_full = np.column_stack((zfemale, zwaist, zhip))

# Create a DataFrame to make the data easier to work with
df_zfemale_full = pd.DataFrame(zfemale_full, columns=['Weight', 'Height', 'BMI', 'Waist Circumference', 'Hip Circumference'])

# Plot the pairplot (scatterplot matrix) for the standardized data
sns.pairplot(df_zfemale_full)
plt.suptitle("Pairplot for Standardized Female Data (Weight, Height, BMI, Waist, Hip)", y=1.02)
plt.show()

# Compute Pearson's and Spearman's correlation coefficients for all pairs of variables
pearson_corr = df_zfemale_full.corr(method='pearson')
spearman_corr = df_zfemale_full.corr(method='spearman')

# Display correlation matrices
print("Pearson's Correlation Matrix:")
print(pearson_corr)

print("\nSpearman's Correlation Matrix:")
print(spearman_corr)
```

1. Body Mass Index (BMI)
Advantages:

Simple and Easy to Use: BMI is calculated using only height and weight, which makes it a quick and easy tool to assess an individual's weight status.
Widely Used: BMI is a common metric in healthcare, fitness assessments, and epidemiological studies.
Good for General Population: It works well at the population level to identify trends related to obesity and underweight.
Disadvantages:

Doesn’t Account for Body Composition: BMI does not distinguish between lean mass (muscle) and fat. Someone with a lot of muscle mass (e.g., athletes) may be classified as overweight or obese, even though they are in good health.
Not Gender-Specific: BMI does not account for gender differences in body composition. For example, women generally have higher body fat than men at the same BMI.
Doesn’t Reflect Fat Distribution: BMI does not show where fat is distributed in the body. A person with a high BMI could be carrying excess weight around the waist (visceral fat), which is associated with higher health risks, or around the hips, which might not be as harmful.
2. Waist-to-Height Ratio (WHtR)
Advantages:

Better Indicator of Health Risks: Waist-to-height ratio is a good predictor of cardiovascular risk and metabolic diseases. It takes into account abdominal fat, which is a strong risk factor for heart disease, diabetes, and other conditions.
More Accurate Than BMI: WHtR provides a better estimate of visceral fat, which is more closely associated with health problems than overall body fat.
Gender-Independent: WHtR is less affected by gender differences in fat distribution, making it a more reliable indicator for both men and women compared to BMI.
Simple and Easy to Calculate: Like BMI, WHtR is easy to compute using basic measurements (waist circumference and height).
Disadvantages:

Does Not Reflect Other Risk Factors: While it’s a good measure for abdominal fat, it does not account for other factors like muscle mass, bone density, or lifestyle.
Requires Waist Measurement: The accuracy of the waist-to-height ratio depends on accurate waist measurement, which can be prone to human error or inconsistencies.
3. Waist-to-Hip Ratio (WHR)
Advantages:

Strong Predictor of Health Risk: WHR is one of the best indicators of body fat distribution, particularly the ratio of abdominal fat (waist) to hip fat. High waist-to-hip ratios are linked to an increased risk of heart disease, diabetes, and stroke.
Helps Identify Visceral Fat: This ratio helps identify individuals with excess visceral fat, which is particularly dangerous for cardiovascular health.
Gender-Specific: WHR can highlight gender differences in fat distribution. For instance, women tend to have more fat in their hips and thighs, while men tend to carry more abdominal fat, making WHR useful in differentiating risk between men and women.
